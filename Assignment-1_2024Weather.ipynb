# <span style="color:Green">P1: Load the dataset.</span>

```python
#read the data from csv file
import csv
with open(r'D:\Algoma\COSC5806001_DataAnalysis\en_climate_daily_ON_6057591_2024_P1D.csv', 'r', encoding="utf-8") as f:
    reader = list(csv.DictReader(f))

#check null value and replace with 0
def replace_empty_with_zero(value):
    if value == "" or value is None:
        return 0
    return value
#print(reader.fieldnames)
```

# <span style="color:Green">P2: What is the overall summary (mean, median, min, max, standard deviation) of temperature, precipitation, and wind speed for the year?</span>

```python
#import required modules
from statistics import mean, median,stdev 
#create empty lists
mean_temperatures =[]
min_temperatures= []
max_temperatures= []
precip = []
windspeed = []
temp_median_ls=[]


#read data of specific columns with name and add to list
for row in reader:
    if 'Mean Temp (°C)' in row:
        mean_temperatures.append(float(replace_empty_with_zero(row['Mean Temp (°C)'].strip())))
    
    if 'Min Temp (°C)' in row:
        min_temperatures.append(float(replace_empty_with_zero(row['Min Temp (°C)'].strip())))
    if 'Max Temp (°C)' in row:
        max_temperatures.append(float(replace_empty_with_zero(row['Max Temp (°C)'].strip())))
    
    if 'Total Precip (mm)' in row:
        precip.append(float(replace_empty_with_zero(row['Total Precip (mm)'].strip())))
            
    if 'Spd of Max Gust (km/h)' in row:
        windspeed.append(float(replace_empty_with_zero(row['Spd of Max Gust (km/h)'].strip())))

#temperature for the year
#median
temp_median_ls.extend(min_temperatures)
temp_median_ls.extend(max_temperatures)
median_temp = median(sorted(temp_median_ls))

#standard deviation
meanTemp = mean(mean_temperatures)
standev = stdev(mean_temperatures,meanTemp)
print('Mean of Temperature is :{}'.format(sum(mean_temperatures)/366))
print('Median of Temperature is :{}'.format(median_temp))
print('Min of Temperature is :{}'.format(sum(min_temperatures)/366))
print('Max of Temperature is :{}'.format(sum(max_temperatures)/366))
print('Standard Deviation of Temperature is :{}'.format(standev))

#precipitation for the year
#standard deviation
meanPrecip = mean(precip)
standev_precip = stdev(precip,meanPrecip)
print('Mean of Precipitation is :{}'.format(mean(precip)))
print('Median of Precipitation is :{}'.format(median(sorted(precip))))
print('Min of Precipitation is :{}'.format(min(precip)))
print('Max of Precipitation is :{}'.format(max(precip)))
print('Standard deviation of Precipitation is :{}'.format(standev_precip))

#wind speed for the year
#standard deviation
meanWindspeed = mean(windspeed)
standev_windspeed = stdev(windspeed,meanWindspeed)
print('Mean of Wind Speed is :{}'.format(mean(windspeed)))
print('Median of Wind Speed is :{}'.format(median(sorted(windspeed))))
print('Min of Wind Speed is :{}'.format(min(windspeed)))
print('Max of Wind Speed is :{}'.format(max(windspeed)))
print('Standard deviation of Wind Speed is :{}'.format(standev_windspeed))
```

# <span style="color:Green">P3: What are the highest and lowest recorded temperatures of the year, and on which dates did they occur?</span>

```python
#Codes of P3 here
# set highest_temp and lowest_temp to temporary val
highest_temp = 0.0
lowest_temp = 0.0
#highest date and lowest date can have more than one
highest_dates = []
lowest_dates = []

for row in reader:
    date, max_temp, min_temp = row['Date/Time'].strip('"'),float(replace_empty_with_zero(row['Max Temp (°C)'].strip('"'))), float(replace_empty_with_zero(row['Min Temp (°C)'].strip('"')))
    
    if max_temp > highest_temp: #check current temp with highest temp 
        highest_temp = max_temp
        highest_dates = [date]  
    elif max_temp == highest_temp:
        highest_dates.append(date)  
        
    if min_temp < lowest_temp: #check current temp with lowest temp 
        lowest_temp = min_temp
        lowest_dates = [date]  
    elif min_temp == lowest_temp:
        lowest_dates.append(date)  
print('Highest Temp is:{} on {}'.format(highest_temp,highest_dates))
print('Lowest Temp is:{} on {}'.format(lowest_temp,lowest_dates))
```

# <span style="color:Green">P4: What is the total annual precipitation (rain + snow)?</span>

```python
#Codes of P4 here
annual_precip = []
for row in reader:
    rain, snow = float(replace_empty_with_zero(row['Total Rain (mm)'].strip('"'))),float(replace_empty_with_zero(row['Total Snow (cm)'].strip('"')))
    snow_mm = snow * 10 if snow else 0 # Convert snow from cm to mm 
    annual_precip.append(rain + snow_mm)
print('Total annual precipitation',sum(annual_precip))
```

# <span style="color:Green">P5: How many days had snowfall, and what was the total snowfall for the year?</span>

```python
#Codes of P5 here
days = 0
total_snowfall = 0.0
for row in reader:
    snowfall = float(replace_empty_with_zero(row['Total Snow (cm)'].strip('"')))
    if snowfall > 0: #check days with snowfall
        days += 1  
        total_snowfall += snowfall
print(f"Snowfall days: {days}")
print(f"Total Snowfall in cm: {round(total_snowfall,2)}")
```

# <span style="color:Green">P6: What is the average temperature for each month?</span>

```python
#Codes of P6 here
monthly_temps = {}
monthly_days = {}
#read temperatures for each month
for i in range(1,13):
    for row in reader:
        month, temp =  int(row['Month'].strip('"')), float(replace_empty_with_zero(row['Mean Temp (°C)'].strip('"')))
        if month in monthly_temps:
            monthly_temps[month]+= temp
            monthly_days[month]+=1
        else:
            monthly_temps[month] = temp
            monthly_days[month]=1

#calculate avg temp for each month
average_monthly_temps = {}
for month, total_temp in monthly_temps.items():
    try:
        average_monthly_temps[month] = total_temp / monthly_days[month]
        print(f"average temp in month {month} is {round(average_monthly_temps[month],2)}")
    except ZeroDivisionError as err:
        print('Error :',err)  
```

# <span style="color:Green">P7: How does the daily temperature vary across different seasons (Winter, Spring, Summer, Fall)? </span>

```python
#Codes of P7 here
seasons = {12: 'Winter', 1: 'Winter', 2: 'Winter',
           3: 'Spring', 4: 'Spring', 5: 'Spring',
           6: 'Summer', 7: 'Summer', 8: 'Summer',
           9: 'Fall', 10: 'Fall', 11: 'Fall'}


season_temps = {}
days_counts = {}
# Calculate total temperatures and get days for each season
for row in reader:
    month, temp =  int(row['Month'].strip('"')), float(replace_empty_with_zero(row['Mean Temp (°C)'].strip('"')))
    if month in seasons:  # Ensure month is in seasons
        season = seasons[month]
        
        if season in season_temps:
            season_temps[season] += temp
            days_counts[season] +=1
        else: 
            season_temps[season] = temp
            days_counts[season] = 1
    else:
        pass

#calculate average of temp in each season
average_season_temps = {}
for season, total_temp in season_temps.items():
    try:
        average_season_temps[season] = total_temp / days_counts[season] if days_counts[season]>0 else 0
        print(f"daily temp in season {season} is {round(average_season_temps[season],2)}")
    except ZeroDivisionError as err:
        print('Error :',err)  
```

# <span style="color:Green">P8: What is the average monthly precipitation, and which month had the highest/lowest precipitation? </span>

```python
#Codes of P8 here
monthly_precip = {}
monthly_days = {}
#group values for each month
for i in range(1,13):
    for row in reader:
        month, precip =  int(row['Month'].strip('"')), float(replace_empty_with_zero(row['Total Precip (mm)'].strip('"')))
        if month in monthly_precip:
            monthly_precip[month]+= precip
            monthly_days[month]+=1
        else:
            monthly_precip[month] = precip
            monthly_days[month]=1

#find the average for monthly precipitation
average_monthly_precip = {}
for month, total_precip in monthly_precip.items():
    try:
        average_monthly_precip[month] = total_precip / monthly_days[month]
        print(f"average precip in month {month} is {round(average_monthly_precip[month],2)}")
    except ZeroDivisionError as err:
        print('Error :',err) 

try:
    # Find the month with maximum average precipitation
    max_month = max(average_monthly_precip, key=average_monthly_precip.get)
    max_precip = average_monthly_precip[max_month]
    
    # Find the month with minimum average precipitation
    min_month = min(average_monthly_precip, key=average_monthly_precip.get)
    min_precip = average_monthly_precip[min_month]
    
    # Print the results
    print(f"Month with highest precipitation: {max_month} with {round(max_precip,2)}")
    print(f"Month with lowest precipitation: {min_month} with {round(min_precip,2)}")
except ValueError as err:
        print('Error :',err) 
       
```

# <span style="color:Green">P9: What is the correlation between temperature and precipitation? The Pearson Correlation formula:
$$r = \frac{\sum(x_i-\overline{x})(y_i-\overline{y})}{\sqrt{\sum{(x_i-\overline{x})}^2\sum{(y_i-\overline{y})}^2}}$$
</span>

```python
#Codes of P9 here
import math
temperature = []
precipitation = []
#get temp and precip values
for row in reader:
    temp, precip =  float(replace_empty_with_zero(row['Mean Temp (°C)'].strip('"'))), float(replace_empty_with_zero(row['Total Precip (mm)'].strip('"')))
    temperature.append(temp)
    precipitation.append(precip)

#calculate mean of temp and precip
try: 
    length = len(temperature) #length is same for temperature and precipitation as 366days so take one
    mean_temp = sum(temperature)/length
    mean_precip = sum(precipitation)/length
except ZeroDivisionError as err:
    print('Error :',err)    

#formula
numerator = 0
temp_deno = 0
precip_deno = 0
for i in range(length):
    numerator += (temperature[i]-mean_temp)*(precipitation[i]-mean_precip)
    temp_deno += (temperature[i]-mean_temp)**2
    precip_deno += (precipitation[i]-mean_precip)**2
    
try:
    correlation = numerator / math.sqrt(temp_deno*precip_deno)
    print("Correlation between temperature and precipitation is {}".format(round(correlation,3)))
except ZeroDivisionError as err:
    print('Error :',err)
```

# <span style="color:Green">P10: How many days recorded extreme weather conditions (e.g., temperature above 30°C or below -20°C)? </span>

```python
#Codes of P10 here
# to store list of dates
highest_dates =[]
lowest_dates = []
for row in reader:
    date, max_temp, min_temp = row['Date/Time'].strip('"'), row['Max Temp (°C)'].strip('"'), row['Min Temp (°C)'].strip('"')
    mintemp = float(replace_empty_with_zero(min_temp))
    if mintemp < -20:  # get dates that fall below -20 temperature
        lowest_dates.append(date)
        
    maxtemp = float(replace_empty_with_zero(max_temp))
    if maxtemp >30:  #get dates that rise above 30 temperature
        highest_dates.append(date)
        
print(f"Days with temperature below -20°C are {lowest_dates}")
print(f"Days with temperature above 30°C are {highest_dates}")
```
