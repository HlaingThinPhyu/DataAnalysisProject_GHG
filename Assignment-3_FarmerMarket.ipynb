# <span style="color:Green">P1: Load the dataset.</span>

```python
#Codes of P1 here
import pandas as pd

# File path to the .xls file
file_path = r'D:\Algoma\COSC5806001_DataAnalysis\Farmers Market Observation-dataset.xlsx'

# Reading the .xls file
data = pd.read_excel(file_path)
# converting into dataframe
df = pd.DataFrame(data)
df
```

# <span style="color:Green">P2: How many unique cities are represented in the dataset? What are the top 5 cities with the most market observations?</span>

```python
#Codes of P2 here
# Get unique values from the city col
unique_cities = df['City'].unique()
# Group by 'cities' and count the occurrences
city_counts = df.groupby('City').size().reset_index(name='count')

# Sort in descending order and get the top 5 cities
top_cities = city_counts.sort_values(by='count', ascending=False).head(5)

print('There are {} unique cities : {}.'.format(unique_cities.size, unique_cities))
print( 'The top 5 cities with the most market observations are ')
print(top_cities.to_string(index= False))  #remove index in printing result
```

# <span style="color:Green">P3: How many markets operate year-round? What is the distribution of market visit months? How many markets are located indoors vs. outdoors?</span>

```python
#Codes of P3 here
#filter spaces in the yearround col and get as str to filter yes value
df['Yearround'] = df['Yearround'].str.strip().astype(str)
yearround_yes = df[df['Yearround'] =='Yes']
print('There are ',len(yearround_yes), 'markets operate year-round.') #print count of result df

#for distributions of market visit month
visit_month = df.groupby('Visit_month').size().reset_index(name='count')
print('The distribution of market visit months are:')
print(visit_month.to_string(index= False)) #remove index in printing result

#indoor vs outdoor markets get each value from location col
df['Location'] = df['Location'].str.strip().astype(str)
loc_indoor = df[df['Location'] == 'Indoors only']
loc_outdoor = df[df['Location'] == 'Outdoors only']
print('There are {} indoor market/s and {} outdoor market/s.'.format(len(loc_indoor),len(loc_outdoor))) #print count of results' df
```

# <span style="color:Green">P4: How many vendors were observed in total? What is the average number of vendors per market? What is the distribution of vendor roles?</span>

```python
#Codes of P4 here
# total no. of vendors 
vendors = df['vendor_ID'].nunique()
print('There are total {} vendors'.format(vendors))

#count vendors by market to get avg number of vendors per mkt
vendors_by_mkt = df.groupby('market_ID')['vendor_ID'].count()
print('Average number of vendors by market: ',vendors_by_mkt.mean().round(2))

#get vendor roles types and clean na from rows, get unique roles and their counts from vendor_role col
vendor_roles = df['vendor_roles'].dropna(axis='rows').value_counts()

print('The distribution of vendor roles are: ')
print(vendor_roles.to_string(header=False))
```

# <span style="color:Green">P5: How many markets have communal sinks available? How many markets allow pets?</span>

```python
#Codes of P5 here
# get rows with communal sink yes
sinks = df[df['communal_sinks'] == 'Yes']
# Count the number of unique market_id values from sinks available df
mkt_wCommunalsinks = sinks['market_ID'].nunique()
print(mkt_wCommunalsinks, 'market/s have communal sinks available.')

# get rows with pet signs yes
pets = df[df['pet_signs'] == 'Yes']
# Count the number of unique market_id values from pets df
mkt_allow_pets = pets['market_ID'].nunique()
print(mkt_allow_pets, 'market/s allow pets.')
```

# <span style="color:Green">P6: What percentage of vendors handle ready-to-eat meats (TCS_rtemeat > 0)? What percentage of vendors sell raw meat?</span>

```python
#Codes of P6 here
#find total unique vendors count to use for percentage calculation
total_vendors = df['vendor_ID'].nunique()

#filter ready_to_eat meat available rows
rtemeat = df[df['TCS_rtemeat'] >0]
# Count the number of unique vendor_ID values from ready_to_eat available and get percentage of vendors from total vendors
vendors_rtemeat = rtemeat['vendor_ID'].nunique()
percent_vendors_rtemeat = (vendors_rtemeat/total_vendors)*100
print('The percentage of vendors handle ready-to-eat meats is: ',round(percent_vendors_rtemeat,2),'%') #round to 2 decimal place

#filter raw meat available rows
rawmeat = df[df['TCS_rawmeat'] >0]
# Count the number of unique vendor_ID values from raw meat available and get percentage of vendors from total vendors
vendors_rawmeat = rawmeat['vendor_ID'].nunique()
percent_vendors_rawmeat = (vendors_rawmeat/total_vendors)*100
print('The percentage of vendors sell raw meat is: ', round(percent_vendors_rawmeat,2),'%') #round to 2 decimal place
```

# <span style="color:Green">P7: How many vendors provide food safety information? How many vendors have visible handwashing stations?</span>

```python
#Codes of P7 here
#filter spaces in the food safety col and get as str to filter 'yes'
df['FS_info'] = df['FS_info'].str.strip().astype(str)
foodsafe_yes = df[df['FS_info'] =='Yes']
print('There are ',len(foodsafe_yes), 'vendors provide food safety information.') #print count of result df

# filter rows with vendors Visible_handwashing_station 
visible_handwash = df[df['Visible_handwashing_station']>0]
print('There are',len(visible_handwash), 'vendors who have visible handwashing stations.') #print count of result df
```

# <span style="color:Green">P8: What is the average number of transactions per vendor? </span>

```python
#Codes of P8 here
#get avg of transactions col, vendor_id col is already unique value so result is avg per vendor
avg_vendor_transactions = df['transactions'].mean()
print('Average number of transactions per vendor:',avg_vendor_transactions.round(2)) #round avg result to 2
```

# <span style="color:Green">P9: What is the distribution of vendor hygiene practices (clean attire)? </span>

```python
#Codes of P9 here
print('Vendors with clean attire hygience practices')
#for clean attire col, get the rows' values and their occurence/count
vendor_cleanattire = df['vendor_cleanattire'].value_counts()
print(vendor_cleanattire.to_string(header= False)) #remove header for user friendly print result
```

# <span style="color:Green">P10: How many vendors use gloves when handling food? What percentage of vendors attempt handwashing? What is the percentage of vendors who touch money and food?</span>

```python
#Codes of P10 here
#find total vendors count
total_vendors = df['vendor_ID'].nunique()

#find handle with gloves and count / vendors ID are already unique so find count of result df
handle_w_gloves = df[df['Handle_gloves'] >0]
print('There are',len(handle_w_gloves), 'vendors use gloves when handling food.')

#filter handwashing_attempts vendors and percentage of them in total
handwashing_attempts = df[df['handwashing_attempts'] >0]
percent_vendors_handwash = round((len(handwashing_attempts)/total_vendors)*100,2)
print(percent_vendors_handwash,'% of vendors attempt handwashing.')

# filter vendors who touch food and cash and percentage of them in total
handle_food_and_cash = df[(df['beh_handle_food'] >0) & (df['beh_cash'] >0) ]
percent_vendor_touch_food_and_cash = round((len(handle_food_and_cash)/total_vendors)*100,2)
print(percent_vendor_touch_food_and_cash,'% of vendors touch money and food.')
```
